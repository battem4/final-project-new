The table presented in the above Url is a knowledge database of disease-symptom associations generated by an automated method based on information in textual discharge summaries of patients at New York Presbyterian Hospital admitted during 2004.  The first column shows the disease, the second the number of discharge summaries containing a positive and current mention of the disease, and the associated symptom. Associations for the 150 most frequent diseases based on these notes were computed and the symptoms are shown ranked based on the strength of association.  The method used the MedLEE natural language processing system to obtain UMLS codes for diseases and symptoms from the notes; then statistical methods based on frequencies and co-occurrences were used to obtain the associations.

Applying classifier to learn diseases from the symptoms

```python
import seaborn as sns
import sklearn
import numpy as np
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
print("Dimensions of x Training set:\n",x_train.shape)
print("Dimensions of y Training set:\n",y_train.shape)

print("Dimensions of x Test set:\n",x_train.shape)
print("Dimensions of y Test set:\n",y_train.shape)
```

```python
mnb = MultinomialNB()
mnb = mnb.fit(x_train, y_train)
```

```python
mnb.score(x_test, y_test)
```
